

**Mean (rata-rata)**

Mean dapat dikatakan sebagai wakil kumpulan data. Menentukan mean dapat dilakukan dengan cara menjumlahkan seluruh nilai data, kemudian membaginya dengan banyaknya data.

* Data Tunggal

  $$
\begin{align*}
\displaystyle\bar x &= \frac{x_{1} + x_{2} + x_{3} + ... + x_{n}}{n} \\
  keterangan &: \\
\bar x &=\ Rata-rata \\
x_{1} + x_{2} + x_{3} + ... + x_{n} &=\ Nilai \  data \\
n &=\ Banyaknya \ data
\end{align*}
$$

* Data kelompok

  *Mean data kelompok artinya data memiliki pengelompokkan dalam interval-interval kelas yang sama panjang.* Untuk mengetahui nilai mediannya dapat ditentukan dengan rumus berikut ini.

$$
\begin{align*}
\displaystyle
\bar x &= \frac{\sum f_{i}x_{i}}{n} \\
keterangan &: \\
\bar x &=\ Rata-rata \\
f_{i} &=\ Frekuensi \\
x_{i} &=\ Nilai \ tengah \\
n &=\ Total \ frekuensi \ atau \ data
\end{align*}
$$



```python
# Mean pada tabel iris
mean = pl.mean()
print('Mean :',mean)
```

**Median (nilai tengah)**

Median adalah nilai tengah yang diambil dari data, yang disusun dengan cara yang berurutan mulai dari kecil hingga terbesar. Median dihitung secara matematis dengan dilambangkan Me. Median adalah nilai di tengah deret nilai yang disusun dengan data urut dari yang kecil sampai besar, misal dalam deret 3,4,7 mediannya adalah 4.
* Data Tunggal<br>
Cara menentukan median pada data tunggal cukup mudah. Anda dapat mengurutkannya dari data terkecil ke data terbesar. Jika banyak data ganjil tentu kita langsung bisa menentukan nilai tengahnya.

  $$
\begin{align*}
\displaystyle
Jumlah \ data \ ganjil &: \\
M_{e} &= \frac{X_{(n -1)}}{2} \\
\end{align*}
$$
Bagaimana jika banyaknya data genap? Pada data tunggal untuk menentukan median yaitu:<br>
Misalkan banyaknya data n, jika n ganjil maka median ditentukan dengan:

  $$
\begin{align*}
\displaystyle
Jumlah \ data \ genap &: \\
M_{e} &= \frac{X_{(\frac{n}{2})}+X_{(\frac{n}{2}+1)}}{2}
\end{align*}
$$

* Data Kelompok

  *Median data kelompok artinya data dikelompokkan dalam interval-interval kelas yang sama panjang.* Untuk mengetahui nilai mediannya dapat ditentukan dengan rumus berikut ini.

$$
\begin{align*}
\displaystyle
M_{e} &= L + c\left( \frac{\frac{1}{2}n - f_{k}}{\sum f_{i}} \right)
\end{align*}
$$
Keterangan:

Me : median

L : tepi bawah kelas median

n : banyaknya data

Fk : frekuensi kumulatif sebelum kelas median

fi : frekuensi kelas median

c : panjang kelas


```python
# Median pada tabel iris
median = pl.median()
print('Median : ',median)
```

**Modus (data yang sering muncul)**

**Modus** adalah sebuah nilai data yang paling sering muncul atau data yang mempunyai nilai frekuensi yang paling tinggi.
* Data Tunggal<br>
  Untuk mencari nilai dari modus pada data tunggal, kita hanya perlu mengurutkan data dari yang terkecil sampai ke yang terbesar atau bisa sebaliknya kemudian kita cek data atau angka berapa yang nilai yang paling banyak muncul.
* Data Kelompok

  Modus data kelompok Modus data kelompok dirumuskan sebagai berikut:

  $$
\begin{align*}
\displaystyle
M_{o} &= L + c\left( \frac{d_{1}}{d_{1} + d_{2}} \right) \\
\end{align*}
$$

M0 : modus data kelompok

L : tepi bawah kelas modus

d1 : selisih frekuensi kelas modus dengan kelas sebelumnya

d2 : selisih frekuensi kelas modus dengan kelas sesudahnya

c : panjang kelas
nilai d1 dan d2 to adalah mutlak (selalu positif)

```python
# Modus pada tabel iris
modus = pl.mode()
print('Modus : ',modus[0])
```

**Kuartil**
  
> *Kuartil adalah nilai-nilai yang membagi data yang telah diurutkan kedalam 4 bagian yang sama besar. Kuartil dinotasikan dengan notasi $Q$. Kuartil terdiri dari 3, yaitu kuartil pertama ($Q_{1}$), kuartil kedua ($Q_{2}$), dan kuartil ketiga ($Q_{3}$).*

>Rumus kuartil data tunggal: 




> Untuk banyaknya data $(n)$ ganjil dan $n+1$ habis dibagi 4.





>$$Q_{1}=x_{(\frac{n+1}{4})}$$

>$$Q_{2}=x_{(\frac{2(n+1)}{4})}$$

>$$Q_{3}=x_{(\frac{3(n+1)}{4})}$$

> Untuk banyaknya data $(n)$ ganjil dan $n+1$ tidak habis dibagi 4.






>$$Q_{1}=\frac{x_{(\frac{(n-1)}{4})}+x_{(\frac{(n+3)}{4})}}{2}$$



>$$Q_{2}={x_{(\frac{2(n+1)}{4})}}$$

>$$Q_{3}=\frac{x_{(\frac{(3n+1)}{4})}+x_{(\frac{(3n+5)}{4})}}{2}$$

> Kuartil untuk banyaknya data $(n)$ genap dan habis dibagi 4.






>$$Q_{1}=\frac{x(\frac{n}{4})+x(\frac{n}{4}+1)}{2}$$



>$$Q_{2}=\frac{x(\frac{n}{2})+x(\frac{n}{2}+1)}{2}$$

>$$Q_{3}=\frac{x(\frac{3n}{4})+x(\frac{3n}{4}+1)}{2}$$

> Kuartil untuk banyaknya data $(n)$ genap dan tidak habis dibagi 4.






>$$Q_{1}=x_{(\frac{n+2}{4})}$$



>$$Q_{2}=\frac{x_{(\frac{n}{2})}+x_{(\frac{n}{2}+1)}}{2}$$

>$$Q_{3}=x_{(\frac{3n+2}{4})}$$

>Rumus kuartil pada data berkelompok






>$$Q_{i}=bb_{Qi}+\frac{\frac{i}{4}n-f_{kjs}}{f_{Qi}}.p$$

>$i$ = kuartil ke-i

>$bb_{Qi}$ = batas bawah kelas kuartil ke-i

>$n$ = banyaknya data

>$f_{kjs}$ = frekuensi kumulatif sebelum kelas kuartil ke-i

>$f_{Qi}$ = frekuensi kelas kuartil ke-i

>$p$ = panjang kelas

```python
# Kuartil pada tabel iris
kuartil = pl.quantile()
print('Kuartil : ',kuartil)
```

**Simpangan Baku**

Standar deviasi dalam statistik, umumnya dipakai untuk menentukan kedekatan sebaran data yang ada di dalam sampel dan seberapa dekat titik data individu dengan mean atau rata-rata nilai dari sampel itu sendiri.

$$
\begin{align*}
\displaystyle
Data \ Populasi &: \\
s &= \sqrt{\frac{\sum_{i=1}^{n}\left( x_{i} - \bar x \right)^{2}}{n}} \\
Data \ Sampel &: \\
s &= \sqrt{\frac{\sum_{i=1}^{n}\left( x_{i} - \bar x \right)^{2}}{n-1}}
\end{align*}
$$

keterangan:
s = simpangan baku

x1 = nilai x ke-i

x = nilai rata-rata data

n = jumlah data

```python
# Simpangan Baku
s = st.pstdev(pl)
print(f'Simpangan Baku : {s}')
```

3. Mendiskripsikan kemiringan/skewned dari setiap variabel (bertipe numerik)

*Kemencengan atau kecondongan (skewness) adalah tingkat ketidaksimetrisan atau kejauhan simetri dari sebuah distribusi. Skewness diartikan sebagai kemiringan distribusi data. Sebuah distribusi yang tidak simetris akan memiliki rata-rata, median, dan modus yang tidak sama besarnya sehingga distribusi akan terkonsentrasi pada salah satu sisi dan kurvanya akan menceng.*

```python
import seaborn as sns
```

```python
skw = c ['sepal_length']
skw.skew()
plt.figure()
# Menampilkan histogram untuk kolom sepal_length menggunakan seaborn
sns.distplot(skw)
plt.show()
```

```python
skw = c ['sepal_width']
skw.skew()
plt.figure()
# Menampilkan histogram untuk kolom sepal_width menggunakan seaborn
sns.distplot(skw)
plt.show()
```

```python
skw = c ['petal_length']
skw.skew()
plt.figure()
# Menampilkan histogram untuk kolom petal_length menggunakan seaborn
sns.distplot(skw)
plt.show()
```

```python
skw = c ['petal_width']
skw.skew()
plt.figure()
# Menampilkan histogram untuk kolom petal_width menggunakan seaborn
sns.distplot(skw)
plt.show()
```
